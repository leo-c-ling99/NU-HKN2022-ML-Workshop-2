```python
import numpy as np
import matplotlib.pyplot as plt
from time import time

# We want to plot the combination of 15 sine waves
start_t = time()
num_points = 1000
x = np.linspace(0, 2*np.pi, num_points)
num_sines = 100
y_components = np.zeros((num_points, num_sines))
for ii in range(num_sines):
    y_components[:, ii] = np.sin(x * 2*np.pi*(ii+1)) 

# Sum different y components
weights = np.array(1/np.arange(1, num_sines+1))
y = np.matmul(y_components, weights)

end_t = time()
print(f'Using Numpy Operations took {end_t - start_t} s')

plt.plot(x, y)
```
